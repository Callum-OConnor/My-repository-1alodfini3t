```python
def points (row):
    if row['FTR'] == 'A' :
        return 3
    if row['FTR'] == 'H' :
        return 3
    if row['FTR'] == 'D' :
        return 1
    return 'Other'
```

```python
df.apply (lambda row: points (row), axis=1)
```

```python
df['points'] = df.apply (lambda row: points(row), axis=1)
```

```python
df[
(df["Date"] > '01/08/2019') &
(df["FTHG"] >= 1) &
(df["HomeTeam"] == "Man United")
]
```

```python
def home_points (row):
    if row['FTR'] == 'A' :
        return 0
    if row['FTR'] == 'H' :
        return 3
    if row['FTR'] == 'D' :
        return 1
    return 'Other'
```

```python
df.apply (lambda row: home_points (row), axis=1)
```

```python
df['home_points'] = df.apply (lambda row: home_points(row), axis=1)
```

```python
df[
(df["Date"] > '01/08/2019') &
(df["FTHG"] >= 1) &
(df["HomeTeam"] == "Man United")
]
```

```python
def away_points (row):
    if row['FTR'] == 'A' :
        return 3
    if row['FTR'] == 'H' :
        return 0
    if row['FTR'] == 'D' :
        return 1
    return 'Other'

df.apply (lambda row: away_points (row), axis=1)

df['away_points'] = df.apply (lambda row: away_points(row), axis=1)
```

```python
df[
(df["HomeTeam"] == "Man United") |
(df["AwayTeam"] == "Man United")
].groupby("FTR")["home_points"].sum()
```
